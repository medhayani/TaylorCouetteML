```python
import os, sys, subprocess, shutil, time
from pathlib import Path
INPUT = Path('/kaggle/input')
CODE_DIR = list(INPUT.rglob('scripts'))[0].parent
WORK = Path('/kaggle/working')
REPO_DIR = WORK / 'TaylorCouetteML'
if REPO_DIR.exists(): shutil.rmtree(REPO_DIR)
shutil.copytree(CODE_DIR, REPO_DIR)
os.chdir(REPO_DIR)
PROFILE_CSV = REPO_DIR / 'data' / 'processed' / 'model_profile_level_dataset.csv'
```

```python
OUT_DIR = WORK / 'runs' / 'ssst_v2'
OUT_DIR.mkdir(parents=True, exist_ok=True)
args = [sys.executable, 'scripts/train_ssst_pro.py',
        '--profile_csv', str(PROFILE_CSV),
        '--out_dir', str(OUT_DIR),
        '--epochs', str(EPOCHS),
        '--batch', str(BATCH_SIZE),
        '--device', 'cuda']
t0 = time.time(); rc = subprocess.call(args, cwd=str(REPO_DIR))
print(f'<<< exit={rc} elapsed={(time.time()-t0)/60:.1f} min')
assert rc == 0
```
